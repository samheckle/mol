```python
import random
evers = []
for x in range(198):
    evers.append(ord('e'))
    evers.append(ord('v'))
    evers.append(ord('e'))
    evers.append(ord('r'))
    
prevChar = '118'

for char in evers:
    if(char == 101 and prevChar != 118):
        print(char, end='  ')
    else:
        print(char, end='')
    rand = random.randrange(100)
    if(rand < 60):
        print(chr(random.randrange(5000)), end='')
    prevChar = char
```

```python
evers = []
for x in range(198):
    evers.append(ord('e'))
    evers.append(ord('v'))
    evers.append(ord('e'))
    evers.append(ord('r'))

for i, char in enumerate(evers):
    if((i % 4) == 0):
        print(char, end=' | ')
    rand = random.randrange(100)
    if(rand < 80):
        for x in range(random.randrange(5)):
            print('\t', end='')
    if(i % 4 != 0):
        print(char, end='')
```
