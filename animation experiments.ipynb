```python
import bezmerizing
from flat import document, shape, rgba
from bezmerizing import Polyline
import numpy as np
from numpy.random import uniform, normal
from IPython.display import SVG, display
def show(page):
    display(SVG(page.svg()))
from random import choice
import ipywidgets as widgets
import random
```

```python
size = 200
d = document(100, 125, 'mm')
page = d.addpage()

for i in range(5):
    curve_figure = shape().stroke(rgba(random.randrange(0, 255), random.randrange(0, 255), random.randrange(0, 255), 255)).width(2)
    pts = []
    width = random.randrange(50, 150)
    height = 5
    steps = 25
    stddev = 0
    for f in range(steps):
        x = ((width / steps) * f) + normal(0, stddev)
        y = normal(0, height)
        pts.append([x, y])
    scribble_poly = Polyline(pts)
    curve = curve_figure.path(scribble_poly.smooth_path().translate(10, (100*(i+.75)/5)))
    page.place(curve)

show(page)
```

```python
float_slider = widgets.IntSlider(
    description="points",
    value=100,
    min=20,
    max=200,
    step=20)
display(float_slider)
```

```python
float_slider = widgets.IntSlider(
    description="points",
    value=100,
    min=20,
    max=200,
    step=20)

def disp_lines(size=100):
    d = document(300, 125, 'mm')
    page = d.addpage()

    for i in range(5):
        curve_figure = shape().stroke(rgba(random.randrange(0, 255), random.randrange(0, 255), random.randrange(0, 255), 255)).width(2)
        pts = []
        width = size
        height = 5
        steps = 25
        stddev = 0
        for f in range(steps):
            x = ((width / steps) * f) + normal(0, stddev)
            y = normal(0, height)
            pts.append([x, y])
        scribble_poly = Polyline(pts)
        curve = curve_figure.path(scribble_poly.smooth_path().translate(10, (100*(i+.75)/5)))
        page.place(curve)
    show(page)
        
widgets.interact(disp_lines, size=float_slider)
```
